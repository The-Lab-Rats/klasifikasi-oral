```python
import shutil
import os
```

```python
def duplicate_images(source_dir, destination_dir, num_duplicates):
    for filename in os.listdir(source_dir):
        if filename.lower().endswith(('.jpg', '.jpeg', '.png', '.gif')):  # Add more image formats if needed
            source_path = os.path.join(source_dir, filename)
            for i in range(1, num_duplicates + 1):
                new_filename = f"{os.path.splitext(filename)[0]}_{i}{os.path.splitext(filename)[1]}"
                destination_path = os.path.join(destination_dir, new_filename)
                shutil.copy2(source_path, destination_path)  # Use copy2 to preserve metadata
                print(f"Image '{filename}' duplicated as '{new_filename}' ({i}/{num_duplicates})")
```

```python
directory = "dataset/lidah leukoplakia//"
num_duplicates = 5  # Number of duplicates for each image

duplicate_images(directory, directory, num_duplicates)
```
